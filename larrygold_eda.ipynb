This notebook explores the classic Iris dataset (150 rows, 5 columns).  
Goal: demonstrate exploratory data analysis using clean data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Plot style
sns.set(style="whitegrid")
pd.set_option("display.max_columns", None)

print("Pandas version:", pd.__version__)
```

```python
df = pd.read_csv("data/raw/iris.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("Missing values:", df.isna().sum().sum())
df.info()
df.describe()

```

```python
df.hist(figsize=(10,6))
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.drop(columns="species").corr(), annot=True, cmap="coolwarm")
plt.show()
```

### Insights
- No missing values; dataset is clean.  
- Setosa species is clearly separable from others by petal length/width.  
- Versicolor and Virginica overlap, but Virginica has generally larger petals.  
- Petal length and petal width are strongly correlated (0.96).
